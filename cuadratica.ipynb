```python
import numpy as np

def der_parcial_0 (xs: list, ys: list) -> tuple[float, float, float]:

    xs_np = np.array(xs)
    ys_np = np.array(ys)

    c_ind = np.sum(ys_np)
    c_2 = np.sum(xs_np**2)
    c_1 = np.sum(xs_np)
    c_0 = len(xs_np)

    return (c_2, c_1, c_0, c_ind)

def der_parcial_1 (xs: list, ys: list) -> tuple[float, float, float]:

    xs_np = np.array(xs)
    ys_np = np.array(ys)

    c_ind = np.sum(ys_np*xs_np)
    c_2 = np.sum(xs_np**3)
    c_1 = np.sum(xs_np**2)
    c_0 = np.sum(xs_np)

    return (c_2, c_1, c_0, c_ind)

def der_parcial_2 (xs: list, ys: list) -> tuple[float, float, float]:

    xs_np = np.array(xs)
    ys_np = np.array(ys)

    c_ind = np.sum(ys_np*xs_np**2)
    c_2 = np.sum(xs_np**4)
    c_1 = np.sum(xs_np**3)
    c_0 = np.sum(xs_np**2)

    return (c_2, c_1, c_0, c_ind)
```

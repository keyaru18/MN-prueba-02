```python
#complejidad eliminacion gaussiana O(n^3)
import numpy as np
def gauss_elimination_sr(x):
    return (x**3 / 3) + (x**2 / 2) - (5*x / 6)
def gauss_elimination_md(x):
    return (x**3 / 3) + (x**2) - (x/3)
def gauss_jordan_sr(x):
    return (x**3 / 2) - (x / 2)
def gauss_jordan_md(x):
    return (x**3 / 2) + (x**2) - (x / 2)
def hybrid_sr(x):
    return (x**3 / 3) + (3/2*x**2) - (5/6*x)
def hybrid_md(x):
    return (x**3 / 3) + (x**2 / 2) - (5/6*x)

#graficar las funciones de complejidad
import matplotlib.pyplot as plt
x = np.linspace(0, 100, 100)
plt.plot(x, gauss_elimination_sr(x), label='Gauss Eliminacion SR')
plt.plot(x, gauss_elimination_md(x), label='Gauss Eliminacion MD')
plt.plot(x, gauss_jordan_sr(x), label='Gauss Jordan SR')
plt.plot(x, gauss_jordan_md(x), label='Gauss Jordan MD')
plt.plot(x, hybrid_sr(x), label='Hibrido SR')
plt.plot(x, hybrid_md(x), label='Hibrido MD')
plt.xlabel('Tamao de la matriz (n)')
plt.ylabel('Operaciones')
plt.title('Complejidad de Algoritmos de Resolucin de Sistemas de Ecuaciones Lineales')
plt.legend()
plt.grid()
plt.show()  
```
